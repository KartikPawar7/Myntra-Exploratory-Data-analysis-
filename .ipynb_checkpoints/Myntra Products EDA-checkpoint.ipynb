```python
brand_data = []

for brand_name in df_prod1['brand_name'].unique():
    
    counts3 = 0
    for i in df_prod1['brand_name']:
        if (i== brand_name):
            counts3+= 1
        
    brand_data.append([brand_name,counts3])

brand_data = pd.DataFrame(brand_data,columns = ['Name of brand','No of Products'])

brand_data.sort_values(by = 'No of Products',ascending=False).head(15)
```

```python
### not mentioned values are basically no info ,can be excluded 
```

```python
pincode = []
state   = []

for i in df_prod1['seller_address']:
    
    if (i != 'Not Mentioned'):
        
        pincode.append(i.split(' ')[-1])
        state.append(i.split(',')[-1].split('-')[0].strip())
        
    else:
        
        pincode.append(np.nan)
        state.append(np.nan)
        
df_prod1['pincode'] = pincode
df_prod1['state']   = state
```

```python
df_prod1.head()
```

```python
df_prod1.groupby('state').size().sort_values(ascending = False).head()
```

```python
df_prod1
```

```python
state_counts  = df_prod1['state'].value_counts()
state_counts = state_counts[:10,]
plt.figure(figsize=(15,5))
sns.barplot(state_counts.index, state_counts.values, alpha=0.8)
plt.title('Top 10 states by producst sales')
plt.ylabel('No of sales', fontsize=10)
plt.xlabel('States', fontsize=10)
plt.show()
```
